```python
# 1. Carga los datos usando tu lector de csv o con pandas. Es recomendable hacerlo con pandas.
import pandas as pd
dataframe = pd.read_csv("avocado.csv")
```

```python
# Identificando el tipo de variables:

dataframe.dtypes

# Todas las variables a excepción de "Date", "type", y "region" son numéricas, 
# ya sean enteras (int) o flotantes/con decimales (float).
# Las variables "Date", "type", y region son variables de texto ("o de tipo string").
```

```python
# 3. Analiza las variables para saber qué representa cada una y en qué rangos se encuentran. 
#    Si la descripción del problema no te lo indica, utiliza el máximo y el mínimo para encontrarlo.

# Obteniendo los valores mínimos y máximos de las variables para saber los rangos de los valores de las mismas:
print("Máximos de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].max())
print("Volumen total:", dataframe["Total Volume"].max())
print("4046:", dataframe["4046"].max())
print("4225:", dataframe["4225"].max())
print("4770:", dataframe["4770"].max())
print("Bolsas totales:", dataframe["Total Bags"].max())
print("Bolsas pequeñas:", dataframe["Small Bags"].max())
print("Bolsas grandes:", dataframe["Large Bags"].max())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].max())
print("Año:", dataframe["year"].max())
print("\n")

print("Mínimos de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].min())
print("Volumen total:", dataframe["Total Volume"].min())
print("4046:", dataframe["4046"].min())
print("4225:", dataframe["4225"].min())
print("4770:", dataframe["4770"].min())
print("Bolsas totales:", dataframe["Total Bags"].min())
print("Bolsas pequeñas:", dataframe["Small Bags"].min())
print("Bolsas grandes:", dataframe["Large Bags"].min())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].min())
print("Año:", dataframe["year"].min())
print("\n")

# Analizando las variables usando algunos ejemplos de filas aleatorias:
print(dataframe.loc[500,:])
print("\n")

print(dataframe.loc[1000,:])
print("\n")

print(dataframe.loc[1500,:])
print("\n")

print(dataframe.loc[2000,:])
print("\n")

print(dataframe.loc[2500,:])
print("\n")

# Se infiere que esta base de datos almacena datos de ventas de bolsas (tanto pequeñas como grandes y muy grandes), así como 
# el promedio total del volumen de las bolsas vendidas en una fecha y región dada, desde el año 2015 hasta el año 2018. Por 
# este motivo, esta base de datos podría haber sido realizada por un negocio que se encargue de vender estas bolsas.

# La variable "Date" almacena las fecha de una jornada de ventas concreta. La variable "Total volume" almacena el volumen total
# de todas las bolsas vendidas en un día determinado, dicho volumen está dado por la cantidad de bolsas pequeñas, grandes y
# muy grandes vendidas en un día específico. No tenemos información suficiente para inferir qué representan las variables
# "4046", "4225" y "4770". La variable "Total Bags" representa la cantidad de bolsas vendidas en un día determinado. La
# variable "Small Bags" representa solo la cantidad de bolsas pequeñas vendidas en un día determinado. La variable "Large
# Bags" representa solo la cantidad de bolsas grandes vendidas en un día determinado. La variable "xLarge Bags" representa la
# cantidad de bolsas muy grandes vendidas en un día en específico. La variable "type" representa el tipo de bolsa vendida. La
# variable "region" representa la localización en la que se realizó la jornada de ventas ese día. Y la variable "year"
# representa el año en el que se llevó a cabo esta jornada de ventas.
```

```python
# 4. Basándose en la media, mediana y desviación estándar de cada variable, ¿Qué conclusiones puedes entregar de los datos?

print("Medias de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].mean())
print("Volumen total:", dataframe["Total Volume"].mean())
print("4046:", dataframe["4046"].mean())
print("4225:", dataframe["4225"].mean())
print("4770:", dataframe["4770"].mean())
print("Bolsas totales:", dataframe["Total Bags"].mean())
print("Bolsas pequeñas:", dataframe["Small Bags"].mean())
print("Bolsas grandes:", dataframe["Large Bags"].mean())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].mean())
print("Año:", dataframe["year"].mean())
print("\n")

print("Medianas de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].median())
print("Volumen total:", dataframe["Total Volume"].median())
print("4046:", dataframe["4046"].median())
print("4225:", dataframe["4225"].median())
print("4770:", dataframe["4770"].median())
print("Bolsas totales:", dataframe["Total Bags"].median())
print("Bolsas pequeñas:", dataframe["Small Bags"].median())
print("Bolsas grandes:", dataframe["Large Bags"].median())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].median())
print("Año:", dataframe["year"].median())
print("\n")

print("Desviación estándar de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].std())
print("Volumen total:", dataframe["Total Volume"].std())
print("4046:", dataframe["4046"].std())
print("4225:", dataframe["4225"].std())
print("4770:", dataframe["4770"].std())
print("Bolsas totales:", dataframe["Total Bags"].std())
print("Bolsas pequeñas:", dataframe["Small Bags"].std())
print("Bolsas grandes:", dataframe["Large Bags"].std())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].std())
print("Año:", dataframe["year"].std())

# Conclusiones de los datos:

# Las desviaciones estándar tanto del volumen total como de los tamaños de las bolsas es muy grande (tomando en cuenta la media 
# de cada variable). Por lo que podemos conlcuir que estos datos van a variar mucho de un año a otro. Por este motivo, si 
# realizáramos una regresión lineal con cada uno de estos datos con respecto al tiempo (en este caso, los años), no tendríamos
# una predicción de qué valores adoptarían en estos datos en un futuro

# Por otra parte, basándonos en la desviación estándar del precio promedio, podemos concluir que si realizáramos una regresión
# lineal con estos datos con respecto al tiempo, tendríamos una mucho mejor predicción de los valores que podrían adoptar en 
# el futuro. La desviación estándar de esta variables es menores a 1, los cuales son valores mucho menores que el resto de 
# las variables. 
```
