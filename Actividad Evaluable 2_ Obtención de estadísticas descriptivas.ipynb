```python
# 3. Analiza las variables para saber quÃ© representa cada una y en quÃ© rangos se encuentran. 
#    Si la descripciÃ³n del problema no te lo indica, utiliza el mÃ¡ximo y el mÃ­nimo para encontrarlo.

# Obteniendo los valores mÃ­nimos y mÃ¡ximos de las variables para saber los rangos de los valores de las mismas:
print("MÃ¡ximos de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].max())
print("Volumen total:", dataframe["Total Volume"].max())
print("4046:", dataframe["4046"].max())
print("4225:", dataframe["4225"].max())
print("4770:", dataframe["4770"].max())
print("Bolsas totales:", dataframe["Total Bags"].max())
print("Bolsas pequeÃ±as:", dataframe["Small Bags"].max())
print("Bolsas grandes:", dataframe["Large Bags"].max())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].max())
print("AÃ±o:", dataframe["year"].max())
print("\n")

print("MÃ­nimos de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].min())
print("Volumen total:", dataframe["Total Volume"].min())
print("4046:", dataframe["4046"].min())
print("4225:", dataframe["4225"].min())
print("4770:", dataframe["4770"].min())
print("Bolsas totales:", dataframe["Total Bags"].min())
print("Bolsas pequeÃ±as:", dataframe["Small Bags"].min())
print("Bolsas grandes:", dataframe["Large Bags"].min())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].min())
print("AÃ±o:", dataframe["year"].min())
print("\n")

# Analizando las variables usando algunos ejemplos de filas aleatorias:
print(dataframe.loc[500,:])
print("\n")

print(dataframe.loc[1000,:])
print("\n")

print(dataframe.loc[1500,:])
print("\n")

print(dataframe.loc[2000,:])
print("\n")

print(dataframe.loc[2500,:])
print("\n")

# Se infiere que esta base de datos almacena datos de ventas de bolsas (tanto pequeÃ±as como grandes y muy grandes), asÃ­ como 
# el promedio total del volumen de las bolsas vendidas en una fecha y regiÃ³n dada, desde el aÃ±o 2015 hasta el aÃ±o 2018. Por 
# este motivo, esta base de datos podrÃ­a haber sido realizada por un negocio que se encargue de vender estas bolsas.

# La variable "Date" almacena las fecha de una jornada de ventas concreta. La variable "Total volume" almacena el volumen total
# de todas las bolsas vendidas en un dÃ­a determinado, dicho volumen estÃ¡ dado por la cantidad de bolsas pequeÃ±as, grandes y
# muy grandes vendidas en un dÃ­a especÃ­fico. No tenemos informaciÃ³n suficiente para inferir quÃ© representan las variables
# "4046", "4225" y "4770". La variable "Total Bags" representa la cantidad de bolsas vendidas en un dÃ­a determinado. La
# variable "Small Bags" representa solo la cantidad de bolsas pequeÃ±as vendidas en un dÃ­a determinado. La variable "Large
# Bags" representa solo la cantidad de bolsas grandes vendidas en un dÃ­a determinado. La variable "xLarge Bags" representa la
# cantidad de bolsas muy grandes vendidas en un dÃ­a en especÃ­fico. La variable "type" representa el tipo de bolsa vendida. La
# variable "region" representa la localizaciÃ³n en la que se realizÃ³ la jornada de ventas ese dÃ­a. Y la variable "year"
# representa el aÃ±o en el que se llevÃ³ a cabo esta jornada de ventas.
```

```python
# 4. BasÃ¡ndose en la media, mediana y desviaciÃ³n estÃ¡ndar de cada variable, Â¿QuÃ© conclusiones puedes entregar de los datos?

print("Medias de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].mean())
print("Volumen total:", dataframe["Total Volume"].mean())
print("4046:", dataframe["4046"].mean())
print("4225:", dataframe["4225"].mean())
print("4770:", dataframe["4770"].mean())
print("Bolsas totales:", dataframe["Total Bags"].mean())
print("Bolsas pequeÃ±as:", dataframe["Small Bags"].mean())
print("Bolsas grandes:", dataframe["Large Bags"].mean())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].mean())
print("AÃ±o:", dataframe["year"].mean())
print("\n")

print("Medianas de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].median())
print("Volumen total:", dataframe["Total Volume"].median())
print("4046:", dataframe["4046"].median())
print("4225:", dataframe["4225"].median())
print("4770:", dataframe["4770"].median())
print("Bolsas totales:", dataframe["Total Bags"].median())
print("Bolsas pequeÃ±as:", dataframe["Small Bags"].median())
print("Bolsas grandes:", dataframe["Large Bags"].median())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].median())
print("AÃ±o:", dataframe["year"].median())
print("\n")

print("DesviaciÃ³n estÃ¡ndar de las variables:")
print("Precio promedio:", dataframe["AveragePrice"].std())
print("Volumen total:", dataframe["Total Volume"].std())
print("4046:", dataframe["4046"].std())
print("4225:", dataframe["4225"].std())
print("4770:", dataframe["4770"].std())
print("Bolsas totales:", dataframe["Total Bags"].std())
print("Bolsas pequeÃ±as:", dataframe["Small Bags"].std())
print("Bolsas grandes:", dataframe["Large Bags"].std())
print("Bolsas muy grandes:", dataframe["XLarge Bags"].std())
print("AÃ±o:", dataframe["year"].std())

# Conclusiones de los datos:

# Las desviaciones estÃ¡ndar tanto del volumen total como de los tamaÃ±os de las bolsas es muy grande (tomando en cuenta la media 
# de cada variable). Por lo que podemos conlcuir que estos datos van a variar mucho de un aÃ±o a otro. Por este motivo, si 
# realizÃ¡ramos una regresiÃ³n lineal con cada uno de estos datos con respecto al tiempo (en este caso, los aÃ±os), no tendrÃ­amos
# una predicciÃ³n de quÃ© valores adoptarÃ­an en estos datos en un futuro

# Por otra parte, basÃ¡ndonos en la desviaciÃ³n estÃ¡ndar del precio promedio, podemos concluir que si realizÃ¡ramos una regresiÃ³n
# lineal con estos datos con respecto al tiempo, tendrÃ­amos una mucho mejor predicciÃ³n de los valores que podrÃ­an adoptar en 
# el futuro. La desviaciÃ³n estÃ¡ndar de esta variables es menores a 1, los cuales son valores mucho menores que el resto de 
# las variables. 
```
